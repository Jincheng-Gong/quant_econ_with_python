```python
import matplotlib.pyplot as plt 
%matplotlib inline 
import numpy as np 
```

```python
thelta_values = np.linspace(0, 2, 10)
x = np.linspace(0, 5, 200)
fig, ax = plt.subplots(figsize=(12,8))

for thelta in thelta_values:
    ax.plot(x, np.cos(np.pi * thelta * x) * np.exp(- x), label=f"$thelta={thelta:.2}$")

ax.legend(loc="upper right")
plt.show()
```
