## Complex Numbers and Euler's Formula
$z = x + iy = re^{i\theta} = r(cos\theta + isin\theta)$

```python
import numpy as np 
import matplotlib.pyplot as plt 
%matplotlib inline
from scipy import *
```

```python
# Abbreviate useful values and functions
pai = np.pi
zero = np.zeros
ones = np.ones
```

```python
# Set parameters
r = 2
theta = pai/3
x = r * np.cos(theta)
x_range = np.linspace(0, x, 1000)
theta_range = np.linspace(0, theta, 1000)

# Plot
fig = plt.figure(figsize=(8, 8)) # 设置正方形画布
ax = plt.subplot(111, projection='polar') # 设置坐标类型为极坐标轴

ax.plot((0, theta), (0, r), marker='o', color='b') # 画两个点，分别是(0, 0)和(theta, r)，画r
ax.plot(zeros(x_range.shape), x_range, color='b') # 画x=1直线，画x
ax.plot(theta_range, x / np.cos(theta_range), color='b') # 画y
ax.plot(theta_range, ones(theta_range.shape) * 0.1, color='r') # 画标志

ax.margins(0) # Let the plot starts at origin

ax.set_title("Trigonometry of complex numbers", va='bottom', fontsize='x-large')

ax.set_rmax(2)
ax.set_rticks((0.5, 1, 1.5, 2))
ax.set_rlabel_position(-88.5)

ax.text(theta, r+0.01, r'$z = x + iy = 1 + \sqrt{3}\, i$')
ax.text(theta+0.2, 1, '$r = 2$')
ax.text(theta-0.2, 0.5, '$x = 1$')
ax.text(0.5, 1.2, r'$y = \sqrt{3}$')
ax.text(0.25, 0.15, r'$\theta = 60^o$')

ax.grid(True)
plt.show()
```

## De Moivre's Theorem
$(r(cos\theta+isin\theta))^n=r^ne^{in\theta}=r^n(cosn\theta+isinn\theta)$
